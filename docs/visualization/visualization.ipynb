### Visualization of strains and acfs

```python
import lal
import numpy as np
import matplotlib.pyplot as plt
from pycbc.types import TimeSeries

plt.rcParams.update({
    'font.family': 'serif',
    'text.usetex': True,
    'axes.labelsize': 22,
    'axes.titlesize': 22,
    'xtick.labelsize': 22,
    'ytick.labelsize': 22,
    'legend.fontsize': 18,
    'xtick.major.size': 8,
    'ytick.major.size': 8,
    'figure.figsize': (10, 6),
    'figure.autolayout': False
})
```

```python
def load_from_txt(txt_path):
    data = np.loadtxt(txt_path)
    times, strain = data[:,0], data[:,1]
    delta_t = times[1] - times[0]
    start_time = times[0]
    strain_series = TimeSeries(strain, delta_t=delta_t, epoch=start_time)
    return strain_series
```

```python
sampling_frequency = 2048.0
start_time = 6665.5
duration_time = 0.1
coalescence_time = 6666
final_mass = 68.20981073643027
delta_t = 28

acf_ET = load_from_txt("../../data/acfs/ET-D_acfs_2048Hz.txt")
strain_ET = load_from_txt("../../data/strains/ET-D_NR0305_strain_2048Hz.txt")
strain_Zero_Noise_221 = load_from_txt("../../data/strains/Zero_Noise_221_2048Hz.txt")
strain_Zero_Noise_222 = load_from_txt("../../data/strains/Zero_Noise_222_2048Hz.txt")
strain_Zero_Noise_223 = load_from_txt("../../data/strains/Zero_Noise_223_2048Hz.txt")

# Strain data (truncated)
rd_analysis_start_time = coalescence_time + final_mass*int(delta_t)*lal.MTSUN_SI
rd_analysis_end_time = rd_analysis_start_time + duration_time
strain_ET_truncated = strain_ET.time_slice(rd_analysis_start_time, rd_analysis_end_time)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(acf_ET.sample_times, acf_ET.numpy(), label='ACF')
plt.title('Autocorrelation Function (ACF)', pad=10)
plt.xlabel('Time (s)')
plt.ylabel('ACF')
plt.legend()
plt.grid(linestyle='--', linewidth=0.8, alpha=0.6)
plt.show()
```

### Visualization of NR0305

```python
plt.figure(figsize=(8, 5))
plt.plot(strain_ET.sample_times, strain_ET.numpy(), label='Strain')
plt.title(f'Strain Data after t0={start_time:.3f}', pad=10)
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.legend()
plt.grid(linestyle='--', linewidth=0.8, alpha=0.6)
plt.show()
```

```python
# Strain data (truncated)
plt.figure(figsize=(8, 5))
plt.plot(strain_ET_truncated.sample_times, strain_ET_truncated.numpy(), label='Strain after t0')
plt.title(f'Strain Data after t0={rd_analysis_start_time:.3f}', pad=10)
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.legend()
plt.grid(linestyle='--', linewidth=0.8, alpha=0.6)
plt.show()
```
